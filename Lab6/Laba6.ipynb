**Обрати один з вказаних методів та стиснути словник і інвертований індекс колекції**




```python
def VB_encode(n):
    bytess = []
    while True:
        bytess.insert(0, n % 128)
        if n < 128:
            break
        n //= 128
    bytess[-1] += 128
    return bytess

def compress(source_path, destination_path, compress_fn=lambda x: [ord(c) for c in x]):
    source_file = open(source_path, "r")
    destination_file = open(destination_path, "wb")
    line = source_file.readline()
    output = b""
    while line:
        split_line = compress_fn(line)
        try:
            for token in split_line:
                encoded = VB_encode(token)
                for byte in encoded:
                    output += (byte).to_bytes(1, "little")
            print("Output:", output)
        except Exception as e:
            print(e, "in token: ", token)

        destination_file.write(output)
        output = b""
        line = source_file.readline()

    source_file.close()
    destination_file.close()
```

```python
def VB_decode(bytestream):
    nums = []
    n = 0
    for idx in range(len(bytestream)):
        if bytestream[idx] < 128:
            n = 128 * n + bytestream[idx]
        else:
            n = 128 * n + (bytestream[idx] - 128)
            nums = n
            n = 0
    return nums

def decompress(source_path, destination_path, decompress_fn=lambda x: chr(x)):
    source_file = open(source_path, "rb")
    destination_file = open(destination_path, "w")
    byte = source_file.read(1)
    bytestream = []
    while byte:
        num = int.from_bytes(byte, "little")
        while not num & 128:
            bytestream.append(num)
            byte = source_file.read(1)
            num = int.from_bytes(byte, "little")
        bytestream.append(num)
        number = VB_decode(bytestream)
        result = decompress_fn(number)
        print(result, end='')
        destination_file.write(str(result))
        bytestream = []
        byte = source_file.read(1)

    source_file.close()
    destination_file.close()
```

```python
compressed = "compressed.dat"
decompressed = "decompressed.txt"
compress("dictOfTerm.txt", compressed)
decompress(compressed, decompressed)
```
