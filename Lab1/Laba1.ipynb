```python
import codecs
import re
import os
import pickle

res = set()
for b in os.listdir('./bookz'):
  file_ = open(os.path.join('./bookz', b), "r")
  f = file_.read()
  w = re.findall("[a-z]+[-']?[a-z]+", f.lower())
  s = set(w)
  res.update(s)
  file_.close()

  print(b)
  print('Words in file: ', len(w))
  print('Uniq: ', len(s))
  print('Uniq %: ', round(len(s)/len(w)*100, 1))
  print()
  
print('Uniq words in all files:', len(res))
```

```python
file_ = open('dictOfTerms.pickle', 'wb')
pickle.dump(res, file_)
file_.close()
file_ = open('dictOfTerms.txt', 'w')
print(' '.join([*res]))
file_.close()
```
